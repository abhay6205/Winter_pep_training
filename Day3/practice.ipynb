```python
movies = {
    "movie-name": ["Dangal", "Bahubali", "Pushpa", "RRR", "KGF"],
    "cast": ["Aamir Khan", "Prabhas", "Allu Arjun", "Ram Charan", "Yash"],
    "total-revenue": [2000, 1800, 1600, 1300, 1200]
}

user_input = input("Enter movie name: ").title()

found = False

for i in range(len(movies["movie-name"])):
    if movies["movie-name"][i].title() == user_input:
        print(
            f"{movies['movie-name'][i].title()} movie casted by "
            f"{movies['cast'][i].title()} and revenue generated by this company is "
            f"{movies['total-revenue'][i]}"
        )
        found = True
        break

if not found:
    print("Movie not found")
```

# Classes

Everything is an object,you can define your own class  to group common actions with data

```python
class MyInt:
    '''docstring for MyInt'''
    def __init__(self,val):
        self.value = val
        
    def __add__(self, other):
        return MyInt(self.value + other)
    
    def __repr__(self):
        return f'MyInt({self.value})'
    
    
```

```python
# study diference between shallow copy and deep copy
# we make shallow copy by just using .copy() method but we make deep copy like this lst1 = lst2
# after making shallow copy, if we make changes in any list then it will make changes in that particular list only
# after making deep copy, if we make changes in any list then it will make chnages in both the list
```

```python
# shallow copy
lst1 = [1,2,3,4]     
lst2 = lst1.copy()
```

```python
lst1, lst2
```

```python
lst2[1] = 100
lst2
```

```python
lst1, lst2
```

```python
# deep copy
lst1 = [1,2,3,4]
lst2 = lst1
```

```python
lst2, lst1
```

```python
lst2[1] = 100
```

```python
lst1, lst2
```

```python
# shallow copy in nested list
lst1 = [[1,2,3,4],[5,6,7.8]]
lst2 = lst1.copy()
```

```python
lst1
```

```python
# using deep copy in nested list
import copy
lst1 = [[1,2,3,4],[5,6,7,8]]
lst2 = copy.deepcopy(lst1)
```

```python
lst2[1][0] = 100
```

```python
lst1, lst2
```

Python Exception Handling

```python
# try-except
try:
    #code block where exception can occur
    a = 1
    b = "s"
    c = a+b
except NameError as exit:        # NameErrorhs
    print("The user has not defined the variable")
except Exception as ex:
    print(ex)
```

```python
# try-except-else
try:
    a = int(input("Enter the number 1: "))
    b = int(input("Enter the number 2: "))
    c=a/b
    d=a*b
    e=a+b
except NameError:
    print("The user has not define the variable")
except ZeroDivisionError:
    print("please provide the number greater than 0")
except TypeError:
    print("Try to make the datatype similar")
except Exception as ex:
    print(ex)
else:
    print(c)
    print(d)
    print(e)
```

```python
#try-else-except with finally
try:
    a = int(input("Enter the number 1: "))
    b = int(input("Enter the number 2: "))
    c=a/b
except NameError:
    print("The user has not define the variable")
except ZeroDivisionError:
    print("please provide the number greater than 0")
except TypeError:
    print("Try to make the datatype similar")
except Exception as ex:
    print(ex)
else:
    print(c)
finally:
    print("The execution has done")
```

```python
# custom exception handling
class Error(Exception):
    pass

class dobException(Error):
    pass

```

```python
year = int(input("Enter the year fo birth: "))
age = 2021 - year
try:
    if age <= 30 & age > 20:
        print("The age is valid. you can apply for the exams")
    else:
        raise dobException
except dobException:
    print("The age is not within the range. you cannot apply for the exams")
```
